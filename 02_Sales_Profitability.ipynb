```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = r"/Users/traceyho/Capstone/Ledger_MasterSales_Clean.csv"

merged = pd.read_csv(path, low_memory=False)

print("âœ… File loaded successfully!")

# Optional: show all column names
print("\nColumn names:")
print(merged.columns.tolist())
```

```python
# --- Step 2: Clean numeric fields ---
def to_num(s):
    s = s.astype(str)
    s = s.str.replace(r"[,\s$%]", "", regex=True)
    s = s.str.replace(r"^\((.*)\)$", r"-\1", regex=True)
    return pd.to_numeric(s, errors="coerce")

merged["qty"] = to_num(merged["Lineitem quantity"])
merged["price"] = to_num(merged["Lineitem price"])
merged["discount"] = to_num(merged.get("Lineitem discount", 0))

# --- Step 3: Compute Revenue, COGS, and Gross Margin ---
# Assume COGS not explicitly given â€” you can refine this once cost data is available.
merged["revenue"] = merged["qty"] * merged["price"] - merged["discount"]

# (optional) if you have inventory fields like â€œValue of Inputsâ€ or â€œValue of Outputsâ€, use them to estimate COGS:
if "Value of Outputs" in merged.columns:
    merged["cogs"] = to_num(merged["Value of Outputs"])
else:
    merged["cogs"] = np.nan  # placeholder

# Gross Profit & Margin
merged["gross_profit"] = merged["revenue"] - merged["cogs"]
merged["gross_margin_pct"] = (merged["gross_profit"] / merged["revenue"]) * 100

# --- Step 4: Add month for analysis ---
merged["Created at"] = pd.to_datetime(merged["Created at"], errors="coerce")
merged["month"] = merged["Created at"].dt.to_period("M").dt.to_timestamp()

# --- Step 5: Summarize ---
month_gm = (
    merged.groupby("month", dropna=True)
    .agg(revenue=("revenue", "sum"),
         cogs=("cogs", "sum"),
         gross_profit=("gross_profit", "sum"))
    .assign(gm_pct=lambda d: (d["gross_profit"] / d["revenue"]) * 100)
    .reset_index()
)

print(month_gm.tail())

# --- Step 6: Visualize ---
plt.figure(figsize=(9,4))
plt.plot(month_gm["month"], month_gm["gm_pct"], marker="o")
plt.title("Gross Margin % by Month")
plt.xlabel("Month")
plt.ylabel("GM %")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# ---- 1) Helpers ----
def exists(col): return col in merged.columns
def to_num(s):
    s = s.astype(str).str.replace(r"[,\s$%]", "", regex=True)
    s = s.str.replace(r"^\((.*)\)$", r"-\1", regex=True)  # (1,234.56) -> -1234.56
    return pd.to_numeric(s, errors="coerce")

# ---- 2) Clean basics we need ----
qty_col   = "Lineitem quantity"
price_col = "Lineitem price"
disc_col  = "Lineitem discount"  # may not exist in all exports

merged["qty"]   = to_num(merged[qty_col]) if exists(qty_col) else np.nan
merged["price"] = to_num(merged[price_col]) if exists(price_col) else np.nan
merged["disc"]  = to_num(merged[disc_col])  if exists(disc_col) else 0.0

# Realized line revenue (pre-refund). You can extend later to allocate refunds if needed.
merged["line_revenue"] = merged["qty"] * merged["price"] - merged["disc"]

# ---- 3) Build/estimate unit cost ----
# Preferred: if you already have a per-line variable cost column, put its exact name here:
per_line_cost_col = None  # e.g., "Product Variable Cost"

if per_line_cost_col and exists(per_line_cost_col):
    merged["unit_cost_est"] = to_num(merged[per_line_cost_col])
else:
    # Fallback: compute SKU average unit cost from inventory data if present:
    # avg_unit_cost = sum(Value of Outputs) / sum(Sales) at SKU level
    if exists("SKU") and exists("Value of Outputs") and exists("Sales"):
        sku_cost = (merged.groupby("SKU", dropna=True)
                          .agg(cogs_value=("Value of Outputs", lambda s: to_num(s).sum(min_count=1)),
                               sales_qty=("Sales", lambda s: to_num(s).sum(min_count=1)))
                          .reset_index())
        sku_cost["avg_unit_cost"] = sku_cost["cogs_value"] / sku_cost["sales_qty"].replace(0, np.nan)
        merged = merged.merge(sku_cost[["SKU","avg_unit_cost"]], on="SKU", how="left")
        merged["unit_cost_est"] = merged["avg_unit_cost"]
    else:
        merged["unit_cost_est"] = np.nan  # no cost available; feel free to join a cost table here

# ---- 4) Aggregate to SKU & compute metrics ----
sku = (merged.groupby(["SKU", "Lineitem name"], dropna=True)
             .agg(units=("qty","sum"),
                  revenue=("line_revenue","sum"),
                  avg_sell_price=("line_revenue", "sum"))  # we'll divide by units next
             .reset_index())

sku["avg_sell_price"] = sku["avg_sell_price"] / sku["units"].replace(0, np.nan)

# bring unit cost estimate (SKU-level) if we computed it above
if "avg_unit_cost" in merged.columns:
    sku = sku.merge(merged[["SKU","avg_unit_cost"]].drop_duplicates("SKU"),
                    on="SKU", how="left")
    sku["unit_cost_est"] = sku["avg_unit_cost"]

# Absolute GP per unit (dollars)
sku["gp_per_unit"] = sku["avg_sell_price"] - sku["unit_cost_est"]

# Optional additional context
sku["total_gross_profit"] = sku["gp_per_unit"] * sku["units"]
sku["margin_pct"] = (sku["gp_per_unit"] / sku["avg_sell_price"]) * 100

# ---- 5) Top 10 by units sold ----
top10_units = (sku.sort_values("units", ascending=False)
                 .head(10)
                 .loc[:, ["SKU","Lineitem name","units","avg_sell_price","unit_cost_est","gp_per_unit","total_gross_profit","margin_pct"]])

# Pretty display
pd.options.display.float_format = lambda x: f"{x:,.2f}"
print("Top 10 SKUs by Units â€” with Absolute Gross Margin per Unit")
print(top10_units)
```

```python
# Units Sold and Revenue by Month

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Helper to clean numerics
def to_num(s):
    s = s.astype(str).str.replace(r"[,\s$%]", "", regex=True)
    s = s.str.replace(r"^\((.*)\)$", r"-\1", regex=True)  # (1,234.56) -> -1234.56
    return pd.to_numeric(s, errors="coerce")

# Dates â†’ month
merged["created_at_ts"] = pd.to_datetime(
    merged["Created at"] if "Created at" in merged.columns else merged["Paid at"],
    errors="coerce"
)
merged["month"] = merged["created_at_ts"].dt.to_period("M").dt.to_timestamp()

# Units & revenue (line level)
merged["qty"]   = to_num(merged["Lineitem quantity"])
merged["price"] = to_num(merged["Lineitem price"])
disc = to_num(merged["Lineitem discount"]) if "Lineitem discount" in merged.columns else 0
merged["revenue"] = merged["qty"] * merged["price"] - disc

# Monthly portfolio table
portfolio = (merged.groupby("month", dropna=True)
                    .agg(units=("qty","sum"),
                         revenue=("revenue","sum"))
                    .reset_index())
```

```python
# Identify Who JOOLAâ€™s Customers Are (Now)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- Prepare customer-level table ---
def to_num(s):
    s = s.astype(str).str.replace(r"[,\s$%]", "", regex=True)
    s = s.str.replace(r"^\((.*)\)$", r"-\1", regex=True)
    return pd.to_numeric(s, errors="coerce")

merged["order_total"]   = to_num(merged.get("Total", np.nan))
merged["line_qty"]      = to_num(merged.get("Lineitem quantity", np.nan))
merged["created_ts"]    = pd.to_datetime(merged.get("Created at", merged.get("Paid at")), errors="coerce")

cust = (merged.groupby("Email", dropna=True)
        .agg(total_spend=("order_total","sum"),
             avg_order_value=("order_total","mean"),
             num_orders=("Email","count"),
             total_units=("line_qty","sum"),
             first_order=("created_ts","min"),
             last_order=("created_ts","max"),
             ship_state=("Shipping Province Name","last"))
        .reset_index())

# --- Visual 1A: AOV distribution ---
plt.figure(figsize=(8,4))
plt.hist(cust["avg_order_value"].dropna(), bins=30)
plt.title("Average Order Value (AOV) Distribution")
plt.xlabel("AOV")
plt.ylabel("Customers")
plt.tight_layout()
plt.show()

# --- Visual 1B: Total spend distribution (trim top 1% for readability) ---
p99 = cust["total_spend"].quantile(0.99)
plt.figure(figsize=(8,4))
plt.hist(cust.loc[cust["total_spend"]<=p99, "total_spend"].dropna(), bins=30)
plt.title("Total Spend per Customer (trimmed at 99th %ile)")
plt.xlabel("Total Spend")
plt.ylabel("Customers")
plt.tight_layout()
plt.show()

# --- Visual 1C: Top 10 shipping states/provinces ---
geo = (cust["ship_state"]
       .value_counts()
       .head(10)
       .iloc[::-1])  # plot horizontal from smallest to largest
plt.figure(figsize=(8,4))
plt.barh(geo.index.astype(str), geo.values)
plt.title("Top 10 Customer Locations")
plt.xlabel("Customers")
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# --- Build simple RFM (same as before) ---
snapshot_date = cust["last_order"].max()
cust["last_order"] = pd.to_datetime(cust["last_order"], errors="coerce")
cust["recency_days"] = (snapshot_date - cust["last_order"]).dt.days
rfm = cust[["recency_days","num_orders","total_spend"]].dropna().copy()

rfm_z = (rfm - rfm.mean()) / rfm.std(ddof=0)

k = 4
km = KMeans(n_clusters=k, random_state=42, n_init=20)
cust.loc[rfm_z.index, "segment"] = km.fit_predict(rfm_z).astype(int)

# --- Segment summary (optional) ---
summary = (cust.groupby("segment")
           .agg(recency_days=("recency_days","mean"),
                freq=("num_orders","mean"),
                monetary=("total_spend","mean"),
                customers=("Email","count"))
           .reset_index())
print(summary)

# === Shared color mapping for all visuals ===
colors = {
    0: "tab:blue",   # Seg 0.0
    1: "tab:orange", # Seg 1.0
    2: "tab:green",  # Seg 2.0
    3: "tab:red"     # Seg 3.0
}

# --- Visual 2A: Segment sizes (bar chart with matching colors + log y-axis) ---
seg_counts = cust["segment"].value_counts().sort_index()

# Custom x labels in segment order 0,1,2,3
segment_labels = [
    "Occasional / Lapsed",   # 0
    "Ultra-loyal / VIP",     # 1
    "Mid-Tier Repeat",       # 2
    "New / One-time"         # 3
]

plt.figure(figsize=(9, 4))

# Use integer positions for bars
x_pos = np.arange(len(seg_counts))

bars = plt.bar(
    x_pos,
    seg_counts.values,
    color=[colors[int(seg)] for seg in seg_counts.index]
)

plt.title("Number of Customers by Segment")
plt.xlabel("Segment")
plt.ylabel("Customers")

# Log scale so tiny Segment 1 is visible
plt.yscale("log")

# Set x tick labels to the descriptive names
plt.xticks(x_pos, segment_labels, rotation=15, ha="right")

# Give some headroom so the top label (25,033) isnâ€™t cut off
plt.ylim(10, seg_counts.max() * 3)

# Add value labels above each bar
for i, val in enumerate(seg_counts.values):
    # Slightly higher offset for all bars; log scale handles it fine
    label_y = val * 1.3
    plt.text(
        i,
        label_y,
        f"{val:,}",
        ha="center",
        va="bottom",
        fontsize=9
    )

plt.tight_layout()
plt.show()
```

```python
# Standardized RFM Heatmap

import numpy as np
import matplotlib.pyplot as plt

# rfm_summary from above
rfm_summary = (cust.groupby("segment")
               .agg(recency_days=("recency_days", "mean"),
                    frequency=("num_orders", "mean"),
                    monetary=("total_spend", "mean"))
               .sort_index())

# Standardize columns (z-scores) along the segments axis
rfm_norm = (rfm_summary - rfm_summary.mean()) / rfm_summary.std(ddof=0)

vals = rfm_norm[["recency_days", "frequency", "monetary"]].values

fig, ax = plt.subplots(figsize=(6, 4))

# Diverging colormap: negatives = below avg, positives = above avg
im = ax.imshow(vals, aspect="auto", cmap="coolwarm")

ax.set_xticks(np.arange(3))
ax.set_xticklabels(["Recency", "Frequency", "Monetary"], rotation=20, ha="right")

ax.set_yticks(np.arange(len(rfm_norm.index)))
ax.set_yticklabels([segment_labels[s] for s in rfm_norm.index])

ax.set_title("Standardized RFM (z-score) by Segment")

# Add colorbar
cbar = plt.colorbar(im, ax=ax)
cbar.set_label("Std devs from segment mean")

# Optional: annotate each cell with numeric value
for i in range(vals.shape[0]):
    for j in range(vals.shape[1]):
        ax.text(j, i, f"{vals[i, j]:.1f}",
                ha="center", va="center", color="black", fontsize=8)

plt.tight_layout()
plt.show()
```

```python
# --- Current revenue share by segment ---
# Map each order line to the customer's segment, then compute revenue share
line_rev = merged[["Email","order_total"]].dropna().copy()
line_rev = line_rev.merge(cust[["Email","segment"]], on="Email", how="left")
rev_share = (line_rev.groupby("segment")["order_total"].sum()
             .pipe(lambda s: s / s.sum())
             .sort_index())

# --- Desired mix (edit these targets to match JOOLA strategy) ---
# Example: grow mid-point buyers & female/family; maintain loyalists
desired_mix = {
    0: 0.25,  # Seg 0 target share
    1: 0.40,  # Seg 1 target share
    2: 0.25,  # Seg 2 target share
    3: 0.10   # Seg 3 target share
}

# Align keys and create aligned arrays
segs = sorted(set(list(rev_share.index) + list(desired_mix.keys())))
current = [float(rev_share.get(s, 0.0)) for s in segs]
target  = [float(desired_mix.get(s, 0.0)) for s in segs]

# --- Visual 3A: Current vs Target revenue mix per segment ---
x = np.arange(len(segs))
width = 0.35
plt.figure(figsize=(9,4))
plt.bar(x - width/2, current, width, label="Current")
plt.bar(x + width/2, target,  width, label="Target")
plt.title("Revenue Mix by Segment: Current vs Target")
plt.xlabel("Segment")
plt.ylabel("Share of Revenue")
plt.xticks(x, [str(s) for s in segs])
plt.legend()
plt.tight_layout()
plt.show()
```
