```python
top_keywords_df = google_data.sort_values(by='Avg. monthly searches', ascending=False).head(10)
display(top_keywords_df)
```

```python
# Select only the date columns and the 'Keyword' column
date_columns = google_data.columns[9:]
columns_to_melt = ['Keyword'] + list(date_columns)
melted_df = top_keywords_df[columns_to_melt]

# Melt the DataFrame to transform the date columns into rows
melted_df = melted_df.melt(id_vars='Keyword', var_name='Date', value_name='Search Volume')

# Convert 'Search Volume' to numeric, coercing errors
melted_df['Search Volume'] = pd.to_numeric(melted_df['Search Volume'], errors='coerce')

# Convert 'Date' to datetime objects
melted_df['Date'] = pd.to_datetime(melted_df['Date'])

# Pivot the melted DataFrame to create the time series format
time_series_df = melted_df.pivot(index='Date', columns='Keyword', values='Search Volume')

# Display the first few rows of the time_series_df
display(time_series_df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
for column in time_series_df.columns:
    plt.plot(time_series_df.index, time_series_df[column], label=column)

plt.xlabel('Date')
plt.ylabel('Search Volume')
plt.title('Monthly Search Volume for Top Keywords (2020-2024)')
plt.legend()
plt.tight_layout()
plt.show()
```
